3) How can you tell the difference between break and continue?

The diffrence between break and continue is that in a loop when the code needs to be terminated from the loop when a particular value comes then break is used, at that particular iteration that particular loop will be terminated, whereas at a particular value , the need of the code is such that it should skip the following iteration and pass the loop to another iteration, continue is used.

below are some examples of break and continue.

```python
for x in range(10):
    print(x)
    if(x==5):
        break
print()
for y in range(10):
    if(y%2==0):
        continue
    print(y)
```

7) Program to find Sum of squares of first n natural numbers.

8) Program to Reverse words in a given String in Python.

```python
string = input("Enter String: ")
rev_str=""
for x in range(len(string)-1,-1,-1):
    rev_str+=string[x]
print(rev_str)
```
